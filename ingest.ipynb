# EDA Limpeza e Transformação

```python
from sqlalchemy import create_engine
```

```python
import pandas as pd

# Carregar o dataset
df = pd.read_csv("train.gz", compression="gzip")

print("Dimensões:", df.shape)
print("\nColunas:", df.columns.tolist())
print("\nTipos de dados:\n", df.dtypes)
print("\nPrimeiras linhas:\n", df.head())
```

```python
print(df.isnull().sum())
```

```python
df['data de referência'] = pd.to_datetime(df['REF_DATE'], utc=True)
```

```python
print(df.describe(include="all"))
```

```python
df.describe()
```

```python
df.describe(include='object')
```

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_percent = (df.isnull().sum()/len(df) * 100).sort_values(ascending=False)
pd.concat([missing_values, missing_percent], axis=1, keys=['Missing Values', 'Percentage'])
```

```python
# Separar colunas numéricas e categóricas
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include='object').columns

# Imputar numéricas com mediana
for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Imputar categóricas com 'Desconhecido'
for col in categorical_cols:
    df[col].fillna('Desconhecido', inplace=True)
```

```python
df.isnull().sum()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols[:10]:  # só primeiras 10 para não poluir
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribuição de {col}')
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols[:10]:
    plt.figure(figsize=(6,4))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Distribuição de {col}')
    plt.show()
```

```python
sns.countplot(x='TARGET', data=df)
plt.title('Distribuição da variável TARGET')
plt.show()

print(df['TARGET'].value_counts(normalize=True))
```

```python
corr = df[numeric_cols].corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Mapa de calor das correlações")
plt.show()
```

```python
for col in numeric_cols[:10]:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot de {col}')
    plt.show()
```
